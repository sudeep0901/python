```python
import nltk
# nltk.download('punkt')
text = "mary had little lamb. Her fleece was white as snow"
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
sents = sent_tokenize(text)
print(sents)
```

```python
words = word_tokenize(text)
print(words)
```

```python
words_token = [word_tokenize(sent) for sent in sents]
print(words_token)
```

```python
#removing stop words (., had, is etcs)
# nltk.download('stopwords')
from nltk.corpus import stopwords
from string  import punctuation
customStopwords = set(stopwords.words('english')+list(punctuation))
print(customStopwords)
```

```python
wordsWostopwords = [word for word in word_tokenize(text) if word not in customStopwords]
print(wordsWostopwords)
```

```python
# identify bigrams
from nltk.collocations import *
biagram_measures = nltk.collocations.BigramAssocMeasures()
```

```python
finder = nltk.BigramCollocationFinder.from_words(wordsWostopwords)
print(finder)
```

```python
sorted(finder.ngram_fd.items())
```

```python
# stemming & POS Tagging
text2 = "Mary closed on closing night when she was in the mood to close."
```

```python
from nltk.stem.lancaster import LancasterStemmer
st=LancasterStemmer()
stemmerWords=[st.stem(word) for word in word_tokenize(text2)]
print(stemmerWords)
```

```python
# nltk.download('averaged_perceptron_tagger')
nltk.pos_tag(word_tokenize(text2))
```

```python
# Disambiguiting word meanings
from nltk.corpus import wordnet as wn
```

```python
# wordnet is like a small thesarus
# nltk.download('wordnet')
for ss in wn.synsets('bass'):
    print(ss, ss.definition())

for et in wn.synsets('eat'):
    print(et, et.definition())
```

```python
# alogrithm for word sense Disambiguation
from nltk.wsd import lesk
sensel = lesk(word_tokenize("Sing in a lower tone, along with the bass"), 'bass')
print(sensel, sensel.definition())
sensel = lesk(word_tokenize("i like the bass fish"), 'bass')
print(sensel, sensel.definition())
sensel = lesk(word_tokenize("i like the bass"), 'bass')
print(sensel, sensel.definition())
```
