```python
def sum1(n):
    final_sum = 0
    for x in range(n +1 ):
        final_sum += x
    
    return final_sum
```

```python
def sum2(n):
    return (n * (n + 1)) / 2
```

```python
%timeit sum1(100)
```

```python
%timeit sum2(100)
```

```python
from math import log
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('bmh')
n = np.linspace(1, 10,100)
# n
labels = ['Constant',"Logarithmic","Linear",'Log Linear', "Quadratic", "Cubic", "Exponential" ]
big_o = [np.ones(n.shape), np.log(n), n, n*np.log(n), n**2, n**3, 2**n]

plt.figure(figsize=(12,10))
plt.ylim(0,50)

for i in range(len(big_o)):
#     print(n, big_o[i])
    plt.plot(n,big_o[i], label=labels[i])
    

plt.legend(loc=0)
plt.ylabel('Relative Runtime')
plt.xlabel('n')
```

```python
# O(1) constant
def func_constant(values):
    print(values[0])
    
func_constant([1, 2, 3, 4, 5, 6, 8]) # does not matter size of list as it always reant first elelment, so constant BIGO
```

```python
# O(n) Linear
def func_linear(ls):
    for val in ls:
        print(val)
    
func_linear([1, 2, 3, 4, 5, 6, 8])
```

```python
# O(n^2) Quadratic
def func_Quadratic(ls):
    for val1 in ls:
        for val2 in ls:
            print(val1, val2)
            
func_Quadratic([1 ,2 , 3, 4])
```

```python
def print_3(lst):
    for i in lst:
        print(i)
    for i in lst:
        print(i)
    for i in lst:
        print(i)

print_3([1, 2, 3]) # it 3(n) Complexity as this run same logic 3 time on lst but still linear only
```

```python
def comp(lst):
    print(lst[0]) # 1
    midpoint = len(lst) // 2 
    for val in lst[:midpoint]: # O n/2
        print(val)
        
    for x in range(10): # 10
        print(x)
        
# O (1 + n/2+ 10) as n grows larger 1 and 10 is irrelevent so O (n / 2) >> O(n)
comp([1, 2, 3, 4, 5, 6 , 7 , 8])
```

```python
# worst / best case
def matcher(lst , match):
    for item in lst:
        if item == match:
            return True
    return False
lst = [1,2, 3, 4,5 ,6,7,8]

matcher(lst,1) # best case O(1)
matcher(list(range(100001)),1000000) # worst case O(n)
```
